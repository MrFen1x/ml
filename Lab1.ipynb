```python
result=[]
sort = data.groupby(['Sex', 'Pclass'])['Age'].mean()
if(sort['male',1]>40):
    result.append('В среднем мужчины 1 класса старше 40 лет')
if(sort['female',1]>40):
    result.append('В среднем женщины 1 класса старше 40 лет')
q3=False
for pclass in [1, 2, 3]:
    men_age = sort['male', pclass]
    women_age = sort['female', pclass]
    if (men_age>women_age):q3=True

if q3 ==True: result.append('Мужчины всех классов в среднем старше, чем женщины того же класса')

sort2 = data.groupby('Pclass')['Age'].mean()
if sort2[1]>sort2[2]>sort2[3]:
    result.append('В среднем, пассажиры первого класса старше, чем пассажиры 2-го класса, которые старше, чем пассажиры 3-го класса.')
    
for q in result:
    print(q)
```
